```python
import pandas as pd

# 1️⃣ Load the datasets
orders = pd.read_csv("olist_orders_dataset.csv")
order_items = pd.read_csv("olist_order_items_dataset.csv")

# 2️⃣ Convert purchase timestamp to datetime
orders['order_purchase_timestamp'] = pd.to_datetime(
    orders['order_purchase_timestamp'], errors='coerce'
)

# 3️⃣ Merge orders with order_items to get price info
df = pd.merge(orders, order_items, on='order_id', how='inner')

# 4️⃣ Create month column
df['month'] = df['order_purchase_timestamp'].dt.to_period('M')

# 5️⃣ Calculate monthly revenue
monthly_revenue = df.groupby('month')['price'].sum().reset_index()

# 6️⃣ Convert month back to datetime for plotting (optional)
monthly_revenue['month'] = monthly_revenue['month'].dt.to_timestamp()

print(monthly_revenue.head())
```

```python
import matplotlib.pyplot as plt

# Remove incomplete months (keep only months before the last month in dataset)
last_month = monthly_revenue['month'].max()
monthly_revenue_filtered = monthly_revenue[monthly_revenue['month'] < last_month]

# Plot without the declining line
plt.figure(figsize=(12, 6))
plt.plot(monthly_revenue_filtered['month'], monthly_revenue_filtered['price'], marker='o')

plt.title("Monthly Revenue Trend", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Revenue", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# --- Data Loading and Preparation ---

# Load all necessary datasets using the correct 'latin1' encoding.
orders_df = pd.read_csv('olist_orders_dataset.csv', encoding='latin1')
order_items_df = pd.read_csv('olist_order_items_dataset.csv', encoding='latin1')
sellers_df = pd.read_csv('olist_sellers_dataset.csv', encoding='latin1')

# Merge the orders and order items dataframes to link sales to sellers and timestamps.
full_orders_df = pd.merge(orders_df, order_items_df, on='order_id')

# Convert the 'order_purchase_timestamp' column to a proper datetime format
# and then extract the month for time-series analysis.
full_orders_df['ds'] = pd.to_datetime(full_orders_df['order_purchase_timestamp'])
full_orders_df['ds'] = full_orders_df['ds'].dt.to_period('M')

# Rename the sales column to 'y', as required by the Prophet library.
full_orders_df.rename(columns={'price': 'y'}, inplace=True)

# Aggregate monthly sales for each seller.
monthly_sales = full_orders_df.groupby(['seller_id', 'ds'], as_index=False)['y'].sum()
monthly_sales['ds'] = monthly_sales['ds'].dt.to_timestamp()

# Identify the top 10 sellers by total revenue.
top_sellers = monthly_sales.groupby('seller_id')['y'].sum().nlargest(10).index
top_sellers_data = monthly_sales[monthly_sales['seller_id'].isin(top_sellers)]

# --- Create a robust seller name lookup dictionary for better titles ---

# Create a mapping from seller_id to Seller_names from the items file.
# Use drop_duplicates() to handle cases where a seller has multiple entries.
seller_names_map = order_items_df.drop_duplicates(subset=['seller_id']).set_index('seller_id').to_dict()

# Create a mapping from seller_id to a descriptive location from the sellers file.
seller_locations_map = sellers_df.set_index('seller_id').apply(
    lambda row: f"{row['seller_city']}, {row}", axis=1
).to_dict()

# Combine the two lookups for the final plot titles.
# This approach prioritizes a specific seller name if it exists, and falls back
# to a city and state combination for a more descriptive title.
seller_info = {}
for seller_id in top_sellers:
    name = seller_names_map.get(seller_id)
    location = seller_locations_map.get(seller_id)
    if pd.notna(name) and name.strip():
        seller_info[seller_id] = f"{name} ({seller_id})"
    elif pd.notna(location) and location.strip():
        seller_info[seller_id] = f"{location} ({seller_id})"
    else:
        seller_info[seller_id] = f"Unknown Seller ({seller_id})"


# --- Time-Series Forecasting and Plotting ---

print("Generating sales forecasts for the top 10 sellers...")

# The following loop processes each of the top-selling merchants individually.
for seller_id in top_sellers:
    # Get the descriptive name for the plot title using our new dictionary.
    plot_title = seller_info.get(seller_id, f"Unknown Seller ({seller_id})")
    
    print(f"\n--- Forecasting for {plot_title} ---")
    
    # Filter the aggregated sales data to get only the current seller's history.
    seller_df = top_sellers_data[top_sellers_data['seller_id'] == seller_id].copy()
    
    # Initialize and fit the Prophet model with the seller's historical data.
    m = Prophet()
    m.fit(seller_df)

    # Create a future dataframe to forecast 6 months ahead.
    future = m.make_future_dataframe(periods=6, freq='M')
    
    # Generate the sales forecast.
    forecast = m.predict(future)
    
    # Display the forecast table for the next 6 months.
    print("Future 6-month sales forecast:")
    print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(6).to_markdown(index=False))

    # Plot the forecast data, showing historical sales, predictions, and uncertainty.
    fig = m.plot(forecast)
    ax = fig.gca()
    ax.set_title(f"Sales Forecast for {plot_title}", fontsize=14)
    ax.set_xlabel("Date", fontsize=12)
    ax.set_ylabel("Projected Sales (yhat)", fontsize=12)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# --------------------
# 1. Load Data
# --------------------
# Specify 'latin1' encoding to handle special characters and prevent UnicodeDecodeError.
try:
    orders_df = pd.read_csv('olist_orders_dataset.csv', encoding='latin1')
    order_items_df = pd.read_csv('olist_order_items_dataset.csv', encoding='latin1')
    sellers_df = pd.read_csv('olist_sellers_dataset.csv', encoding='latin1')
except FileNotFoundError:
    print("One or more CSV files not found. Please ensure they are in the same directory.")
    exit()

# --------------------
# 2. Data Preprocessing
# --------------------

# Merge the dataframes to link orders, order items, and sellers.
full_orders_df = pd.merge(orders_df, order_items_df, on='order_id')
full_orders_df = pd.merge(full_orders_df, sellers_df, on='seller_id')

# Convert timestamp column to datetime objects.
full_orders_df['order_purchase_timestamp'] = pd.to_datetime(full_orders_df['order_purchase_timestamp'])

# Aggregate revenue monthly for each seller.
monthly_revenue_df = full_orders_df.groupby([
    'seller_id', 
    'Seller_names',
    pd.Grouper(key='order_purchase_timestamp', freq='M')
])['price'].sum().reset_index()

# Rename columns for Prophet model.
monthly_revenue_df.rename(columns={'order_purchase_timestamp': 'ds', 'price': 'y'}, inplace=True)

# --------------------
# 3. Identify and Map Top Sellers
# --------------------

# Find the top 10 sellers by total revenue.
top_sellers_df = monthly_revenue_df.groupby('seller_id')['y'].sum().nlargest(10).reset_index()
top_sellers = top_sellers_df['seller_id'].tolist()

# Create a mapping for seller names to use in titles.
# Use Seller_names if available, otherwise use a default name.
seller_name_mapping = full_orders_df.set_index('seller_id')['Seller_names'].to_dict()

# --------------------
# 4. Forecasting and Visualization
# --------------------
plt.style.use('fivethirtyeight')
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(15, 20), sharex=False, sharey=False)
fig.suptitle('Monthly Revenue Forecasts for Top 10 Sellers', fontsize=20, y=1.02)
axes = axes.flatten()

for i, seller_id in enumerate(top_sellers):
    # Filter data for the current seller.
    seller_data = monthly_revenue_df[monthly_revenue_df['seller_id'] == seller_id]
    
    # Check if seller_data is not empty and has enough data points.
    if not seller_data.empty and len(seller_data) > 1:
        # Create a new Prophet model for each seller.
        m = Prophet(seasonality_mode='multiplicative',
                    changepoint_prior_scale=0.05,
                    n_changepoints=25)
        m.fit(seller_data)
        
        # Create a future dataframe for 6 months.
        future = m.make_future_dataframe(periods=6, freq='M')
        
        # Generate the forecast.
        forecast = m.predict(future)
        
        # Get the seller name from the mapping or use a default.
        seller_name = seller_name_mapping.get(seller_id, f"Unknown Seller ({seller_id[:8]})")

        # Plot the forecast on the appropriate subplot.
        m.plot(forecast, ax=axes[i])
        m.plot(forecast, ax=axes[i], uncertainty=True)
        axes[i].set_title(f'Seller: {seller_name}', fontsize=12)
        axes[i].set_xlabel('Date')
        axes[i].set_ylabel('Revenue')
    else:
        # Handle sellers with insufficient data.
        axes[i].set_title(f'Seller: Not Enough Data', fontsize=12)
        axes[i].set_xlabel('Date')
        axes[i].set_ylabel('Revenue')
        axes[i].plot([], []) # Plot an empty line to avoid errors.

# Adjust layout and display plots.
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# --------------------
# 5. Display Forecast Data
# --------------------
print("\nForecast Data Tables:")
for seller_id in top_sellers:
    seller_data = monthly_revenue_df[monthly_revenue_df['seller_id'] == seller_id]
    if not seller_data.empty and len(seller_data) > 1:
        m = Prophet(seasonality_mode='multiplicative', changepoint_prior_scale=0.05)
        m.fit(seller_data)
        future = m.make_future_dataframe(periods=6, freq='M')
        forecast = m.predict(future)
        
        seller_name = seller_name_mapping.get(seller_id, f"Unknown Seller ({seller_id[:8]})")
        print(f"\nForecast for Seller: {seller_name}")
        print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(6).to_string())
```

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# --- 1. LOAD AND MERGE THE DATASETS ---
# Load all necessary datasets with the correct 'latin1' encoding.
try:
    orders_df = pd.read_csv('olist_orders_dataset.csv', encoding='latin1')
    order_items_df = pd.read_csv('olist_order_items_dataset.csv', encoding='latin1')
    sellers_df = pd.read_csv('olist_sellers_dataset.csv', encoding='latin1')

    # Merge the dataframes to link orders, items, and sellers
    # Start with order_items_df, as it connects both orders and sellers
    merged_df = pd.merge(order_items_df, orders_df, on='order_id')
    merged_df = pd.merge(merged_df, sellers_df, on='seller_id')

except FileNotFoundError as e:
    print(f"Error: One of the files was not found. Please ensure all three files are present: {e}")
    exit()
except Exception as e:
    print(f"An unexpected error occurred during file loading or merging: {e}")
    exit()

# --- 2. DATA CLEANING AND PREPARATION ---
# Convert the order purchase timestamp to a datetime object
merged_df['order_purchase_timestamp'] = pd.to_datetime(merged_df['order_purchase_timestamp'])

# Calculate the total revenue (price + freight) for each item
merged_df['Revenue'] = merged_df['price'] + merged_df['freight_value']

# Group the data by seller and month to get monthly revenue for each seller
monthly_revenue_df = merged_df.groupby([pd.Grouper(key='order_purchase_timestamp', freq='M'), 'Seller_names'])['Revenue'].sum().reset_index()
monthly_revenue_df.rename(columns={'order_purchase_timestamp': 'Date', 'seller_id': 'Seller_names'}, inplace=True)

# --- 3. RANK SELLERS BY TOTAL REVENUE ---
total_revenue_by_seller = monthly_revenue_df.groupby('Seller_names')['Revenue'].sum()
top_10_sellers = total_revenue_by_seller.sort_values(ascending=False).head(10).index

print("Top 10 Sellers by Total Revenue:")
print("-" * 50)
for seller in top_10_sellers:
    revenue = total_revenue_by_seller.loc[seller]
    print(f"{seller:<35}: ${revenue:,.2f}")

# --- 4. FUTURE FORECASTING AND VISUALIZATION ---
forecast_horizon = 6

print("\n" + "="*50)
print(f"Generating forecasts and visualizations for the next {forecast_horizon} months:")
print("="*50)

for seller in top_10_sellers:
    seller_data = monthly_revenue_df[monthly_revenue_df['Seller_names'] == seller].set_index('Date')['Revenue']
    seller_data = seller_data.asfreq('M', fill_value=0) # Resample to ensure monthly frequency
    
    try:
        model = sm.tsa.ARIMA(seller_data, order=(1,1,1))
        results = model.fit()
        forecast = results.forecast(steps=forecast_horizon)
        
        forecast_dates = pd.date_range(start=seller_data.index[-1], periods=forecast_horizon + 1, freq='M')[1:]
        
        plt.figure(figsize=(10, 6))
        plt.plot(seller_data.index, seller_data.values, label='Historical Revenue', color='blue')
        plt.plot(forecast_dates, forecast.values, label='Forecasted Revenue', color='red', linestyle='--')
        
        plt.title(f'Historical and Forecasted Revenue for {seller}', fontsize=16)
        plt.xlabel('Date', fontsize=12)
        plt.ylabel('Revenue', fontsize=12)
        plt.legend()
        plt.grid(True)
        plt.xticks(rotation=45)
        plt.tight_layout()
        
        plt.show()
        
    except Exception as e:
        print(f"Could not fit ARIMA model for {seller} or plot the data. Error: {e}")
        plt.close()
```
